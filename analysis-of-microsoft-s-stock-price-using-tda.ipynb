```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import yfinance as yf

# Fetch Microsoft stock data
df_msft = yf.download('MSFT', start='2020-01-01', end='2024-12-16',progress=False)
df_msft.max()
```

```python
df_msft
```

```python
df_msft.min()
```

```python
highest_price_date = df_msft['High'].idxmax()
highest_price_value = df_msft['High'].max()

print(f"The stock price was highest on {highest_price_date}, with a value of {highest_price_value:.2f}.")

lowest_price_date = df_msft['High'].idxmin()
lowest_price_value = df_msft['High'].min()

print(f"The stock price was lowest on {lowest_price_date}, with a value of {lowest_price_value:.2f}.")





```
