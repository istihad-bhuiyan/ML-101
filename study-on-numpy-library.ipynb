How the axis Parameter Works:

axis=0 :

Concatenation is performed along rows (vertically).
The number of rows increases, but the columns remain the same.
In this case:
arr1 has 2 rows and 2 columns.
arr2 has 2 rows and 2 columns.
The result arr has 4 rows and 2 columns.

axis=1:

Concatenation is performed along columns (horizontally).
The number of columns increases, but the rows remain the same.

If the value at an index is True that element is contained in the filtered array, if the value at that index is False that element is excluded from the filtered array.

```python
import numpy as np

arr = np.array([41, 42, 43, 44])


filter_arr = []


for element in arr:
 
  if element > 42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)

newarr = arr[filter_arr]

print(filter_arr)
print(newarr)
```

**Discrete Distribution:**> The distribution is defined at separate set of events, e.g. a coin toss's result is discrete as it can be only head or tails whereas height of people is continuous as it can be 170, 170.1, 170.11 and so on.

```python
from numpy import random as rnd

x = rnd.binomial(n=1000 ,p=0.5, size=10)

print(x)
```

n - number of trials.

p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

size - The shape of the returned array.

Difference Between Logistic and Normal Distribution

Both distributions are near identical, but logistic distribution has more area under the tails, meaning it represents more possibility of occurrence of an event further away from mean.

For higher value of scale (standard deviation) the normal and logistic distributions are near identical apart from the peak.

<h4>A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).</h4>
<h4>A smaller a (e.g. a<2) leads to a heavier tail, meaning extreme values (large outliers) are more likely.
A larger a (e.g., >2) results in a lighter tail, meaning values concentrate closer to the minimum.</h5>

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.pareto(a=10, size=(2, 3))
sns.kdeplot(random.pareto(a=2,size=1000))
print(x)
plt.show()
sns.histplot(random.pareto(a=2,size=1000))
plt.show()
```

<h4>Zipf's Law: In a collection, the nth common term is 1/n times of the most common term. E.g. the 5th most common word in English occurs nearly 1/5 times as often as the most common word.
</h4>

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=1000)
sns.kdeplot(x[x<20])
# print(x)
plt.show()
sns.histplot(x[x<20])
plt.show()
```

```python
import numpy as np

def myadd(x, y,z):
  return 5*x+2*y-8*z

myadd = np.frompyfunc(myadd, 3, 1)

print(myadd([1, 2, 3, 4], [5, 6, 7, 8],[5,4,3,4]))
```
