<h2> Time Resampling


```python
df_tesla.head()
```

```python
df_tesla.resample(rule='YE')   #'YE' (Year-End)
#Groups the data into periods ending on the last day of each calendar year 
```

```python
df_tesla.resample(rule='YE').min()
```

```python
df_tesla.resample(rule='YE').max()
```

```python
type(df_tesla.resample(rule='YE').max())
```

```python
import pandas as pd
df_tesla.resample(rule='YE').max()['Open'].plot()
```

```python
df_tesla.resample(rule='QS').min() #Quarterly
```

```python
df_tesla.resample(rule='QS').max()['High'].plot()
```

<h4> Quarterly plotting gives better visualization.

```python
df_tesla.resample(rule='MS').max()['High'].plot() #Monthly
```

```python
df_tesla.resample(rule='D').max()['High'].plot() #Daily
```

```python
df_tesla['Open'].resample(rule='YE').min().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='QS').min().plot(kind='bar')
```

```python
df_tesla['High'].rolling(11).max().head(20)
```

<h4>This will compute the maximum of the previous 11 days for each day in the DataFrame.</h4>

```python
df_tesla['Thirty_days_rolling'] = df_tesla['Open'].rolling(window=30).mean()
df_tesla['Thirty_days_rolling'] = df_tesla['Thirty_days_rolling'].fillna(method='ffill')

df_tesla.head(90)
```

```python
df_tesla[['Open','Thirty_days_rolling']].plot(figsize=(15,10))
```
